# Newton method for optimization

## Objective function:

$f = 4(x_1 + 1)^2 + 2(x_2 - 1)^2 + x_1 + x_2 + 10$

### gradient of objective function:

$g = (8x_1 + 9,4x_2 - 3)$

### Hessian of objective function:

$H = ([8,0],[0,4])$

## code:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def gradient(x):
    return np.array([8*x[0]+9,4*x[1]-3])

H = np.array([[8,0],[0,4]])

X1 = np.arange(-2,2.1,0.1)

Y1 = np.arange(-2,2.1,0.1)

[x1,y1]=np.meshgrid(X1,Y1)

f = 4*(x1+1)**2 + 2*(y1-1)**2 + x1 + y1 +10

plt.contour(x1,y1,f,20)

def newton(x0):
    print("初始点为",x0)
    imax =100
    record = np.zeros((2,imax))
    record[:,0]=x0
    x = x0
    i = 1
    alpha = 1
    epsilon = 1
    
    while i < imax and epsilon > 10**(-5):
        decl = -np.dot(np.linalg.inv(H),gradient(x))
        x0 = x
        x = x + alpha*decl
        record[:,i] = x
        epsilon = sum((x-x0)**2)
        print("第{0}次迭代点为{1}".format(i,x),'\n')
        i = i + 1
    record = record[:,0:i]
    return record

x0 = np.zeros((2))
record = newton(x0)

plt.plot(record[0,:],record[1,:],'g*',record[0,:],record[1,:])
plt.show()
```
